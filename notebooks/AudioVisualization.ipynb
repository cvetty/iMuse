```python
import numpy as np
import matplotlib.pyplot as plt
import librosa as lb
import librosa.display
```

### Audio Visualization Explore

```python
FRAME_LENGTH = 512

audio, sr = lb.load(r"D:\Projects\SCHOOL\IMuse\data\music\others\Q2\MT0007379559.mp3", duration=3)
```

```python
lb.display.waveshow(audio, sr)
```

```python
spec = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=128,fmax=8000, hop_length=FRAME_LENGTH)
specm=np.mean(spec,axis=0)
gradm=np.gradient(specm)
gradm=gradm/np.max(gradm)
gradm = gradm.clip(min=0)

### gradm -> maps to style_boost_coeff

specm=(specm-np.min(specm))/np.ptp(specm)
```

```python
plt.plot(specm + gradm)
```
